## Testes de desempenho para os tipos de dados JSON e JSONB

### Testes de desempenho para o tipo de dados text[]

Para propósitos de testes de desempenho adicionais, foi criada uma nova tabela *textarrtest* para testar e comparar os desempenhos de acesso aos dados armazenados em três colunas do tipo *text[]* (*text, bold, nonbold*)

Para criar a tabela *textarrtest*, foi utilizado o seguinte script, a partir das tabelas existentes:

```plsql
SELECT
	t.id,
	ARRAY(SELECT jsonb_array_elements_text(t.interview->'text')) AS text, 
	ARRAY(SELECT jsonb_array_elements_text(t.interview->'bold')) AS bold, 
	ARRAY(SELECT jsonb_array_elements_text(t.interview->'nonbold')) AS nonbold
	INTO textarrtest
	FROM jsonbtest t;
```

````python
# Preparing for N queries
times = []
labels = []

N = 1000
````

````python
# Accessing  an entire tuple of JSONBTEST
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONBTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSONB entire row")

# Accessing  an entire tuple of JSONTEST
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSON entire row")

# Accessing  an entire tuple of TEXT[]
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM TEXTARRTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("TEXT[] entire row")
````

````python
# Accessing an entire JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT INTERVIEW FROM JSONBTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSONB entire object")

# Accessing an entire JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT INTERVIEW FROM JSONTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSON entire object")

# Accessing an entire TEXT[] data
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT TEXT, BOLD, NONBOLD FROM TEXTARRTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("TEXT[] entire object")
````

````python
# Accessing 'bold' from a JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT INTERVIEW->'bold' FROM JSONBTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSONB 'bold'")

# Accessing 'bold' from a JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT INTERVIEW->'bold' FROM JSONTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSON 'bold'")

# Accessing 'bold' from a TEXT[]
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT BOLD FROM TEXTARRTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("TEXT[] 'bold'")
````

````python
# Accessing 'nonbold' from a JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT INTERVIEW->'nonbold' FROM JSONBTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSONB 'nonbold'")

# Accessing 'nonbold' from a JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT INTERVIEW->'nonbold' FROM JSONTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("JSON 'nonbold'")

# Accessing 'nonbold' from a TEXT[]
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT NONBOLD FROM TEXTARRTEST WHERE ID = %(int)s;""", {'int': randint(1, 10000)})
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times.append(aux)
labels.append("TEXT[] 'nonbold'")
````

````python
# Accessing entire table JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONBTEST;""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[2][0])[0]))
times.append(aux)
labels.append("JSONB *")

# Accessing entire table JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONTEST;""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[2][0])[0]))
times.append(aux)
labels.append("JSON *")

# Accessing entire table TEXT[]
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM TEXTARRTEST;""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[2][0])[0]))
times.append(aux)
labels.append("TEXT[] *")
````

````python
print('Average access times (ms): ')
for i in range(len(times)):
    print(labels[i], '%.5f' % (sum(times[i])/N),sep='\t\t')
````

````python
# Graph comparison

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times[0:3], y=labels[0:3])
plt.xlim(0, 0.1)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[]")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times[3:6], y=labels[3:6])
plt.xlim(0, 0.1)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[]")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times[6:9], y=labels[6:9])
plt.xlim(0, 1.4)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[]")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times[9:12], y=labels[9:12])
plt.xlim(0, 1.4)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[]")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times[12:15], y=labels[12:15])
plt.xlim(0, 1.4)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[]")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times[15:18], y=labels[15:18])
plt.xlim(0, 5)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[]")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()
````

## Testes de desempenho para busca por termos

Para exemplificar, as consultas a seguir retornam todas as tuplas cuja a *string* "Londrina" é existente nos elementos *text*, *bold* e *nonbold*. Neste caso, todas as tuplas das tabelas de testes são retornadas após a verificação.

````python
# Preparing for N queries
times2 = []
labels2 = []

N = 10
````

````python
# TERM SEARCH 'text' JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONBTEST JBT WHERE INTERVIEW->>'text' LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("JSONB 'text' TERM SEARCH")

# TERM SEARCH 'text' JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONTEST JT WHERE INTERVIEW->>'text' LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("JSON 'text' TERM SEARCH")

# TERM SEARCH 'text' TEXTARRTEST
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM TEXTARRTEST WHERE array_to_string(TEXT, '|') LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("TEXT[] 'text' TERM SEARCH")
````

````python
# TERM SEARCH 'bold' JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONBTEST JBT WHERE INTERVIEW->>'bold' LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("JSONB 'bold' TERM SEARCH")

# TERM SEARCH 'bold' JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONTEST JT WHERE INTERVIEW->>'bold' LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("JSON 'bold' TERM SEARCH")

# TERM SEARCH 'bold' TEXTARRTEST
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM TEXTARRTEST WHERE array_to_string(BOLD, '|') LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("TEXT[] 'bold' TERM SEARCH")
````

````python
# TERM SEARCH 'nonbold' JSONB
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONBTEST JBT WHERE INTERVIEW->>'nonbold' LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("JSONB 'nonbold' TERM SEARCH")

# TERM SEARCH 'nonbold' JSON
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM JSONTEST JT WHERE INTERVIEW->>'nonbold' LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("JSON 'nonbold' TERM SEARCH")

# TERM SEARCH 'nonbold' TEXTARRTEST
cur = conn.cursor()
aux = []
for i in range(N):
    cur.execute("""EXPLAIN (ANALYZE TRUE, TIMING FALSE) SELECT * FROM TEXTARRTEST WHERE array_to_string(NONBOLD, '|') LIKE '%Londrina%';""")
    aux.append(float(re.findall("\d+\.\d+", cur.fetchall()[3][0])[0]))
times2.append(aux)
labels2.append("TEXT[] 'nonbold' TERM SEARCH")
````

````python
print('Average access times (ms): ')
for i in range(len(times2)):
    print(labels2[i], '%.5f' % (sum(times2[i])/N),sep='\t\t')
````

````python
# Graph comparison

sns.set()
plt.figure(figsize=(16, 9))
sns.boxplot(x=times2[0:3], y=labels2[0:3])
plt.xlim(0, 10000)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[], com buscas por termos")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()
sns.set()

plt.figure(figsize=(16, 9))
sns.boxplot(x=times2[3:6], y=labels2[3:6])
plt.xlim(0, 10000)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[], com buscas por termos")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()

plt.figure(figsize=(16, 9))
sns.boxplot(x=times2[6:9], y=labels2[6:9])
plt.xlim(0, 10000)
plt.title("Comparação de desempenho para queries em dados JSON, JSONB e TEXT[], com buscas por termos")
plt.ylabel("Consulta")
plt.xlabel("Tempo de execução (ms)")
plt.plot()
````

## Conclusões

Pelos resultados observados percebe-se que, para a maioria das consultas utilizadas o tipo de dados "JSON" apresentou um tempo de acesso maior que os demais, assim como o espaço ocupado em disco para os dados utilizados. Sendo assim, entre os dois tipos de dados para a manipulação de JSON, o "JSONB" aparenta ser uma escolha superior.

No entanto, comparando os tipos "JSONB" e "TEXT[]", para a maioria das consultas o tipo "TEXT[]" obteve um desempenho maior, sem muitas alterações por parte da aplicação (como pode ser observado no código-fonte).

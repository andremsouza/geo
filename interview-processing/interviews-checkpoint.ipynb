```python
import docx2json
import json
from nltk.stem.snowball import SnowballStemmer
import nltk
import string
```

Import file text

```python
fp = "/home/andre/Documents/ic/sample-files/Camila Ferezin do Amarante ( Ginastica ritmica ).docx"
```

```python
doc = json.loads(docx2json.convert(fp))
```

```python
interview = '\n'.join(doc['nonbold'])
```

Remove punctuation

```python
interview_nopunct = interview.translate(str.maketrans('', '', string.punctuation+string.digits))
```

Tokenize text, remove stopwords

```python
interview_split = interview_nopunct.lower().split()
```

```python
tokens = set(interview_split)-set(nltk.corpus.stopwords.words('portuguese'))
```

Create bag-of-words and stemming

```python
stemmer = SnowballStemmer('portuguese')
bag, bag_stemmed = {}, {}

for token in tokens:
    bag[token] = interview_split.count(token)
    bag_stemmed[stemmer.stem(token)] = bag_stemmed.get(stemmer.stem(token), 0) + bag[token]
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud().generate_from_frequencies(bag_stemmed)
```

```python
# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
